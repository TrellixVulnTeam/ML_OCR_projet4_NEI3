# Suppression des colonnes inutiles 

```python
customers = customers.drop(columns = ['customer_zip_code_prefix'])
order_items = order_items.drop(columns = ['shipping_limit_date'])
order_payments = order_payments.drop(columns = ['payment_sequential'])
orders = orders.drop(columns = ['order_approved_at',
                               'order_delivered_carrier_date',
                               'order_delivered_customer_date',
                               'order_estimated_delivery_date'])
```

# Fusion des databases products et product_category

Utilisation de l'appelation anglaise des produits dans la database

```python
df1 = pd.merge(products , product_category, on = 'product_category_name', how = 'left')
df1 = df1.drop(columns = ['product_category_name']) 
df1.head()
```

# Merge des datasets orders, customers, order_items, order_payments et sellers

```python
df_merge = pd.merge(orders,customers, on = 'customer_id')
df_merge = df_merge.merge(order_items, on = 'order_id')
df_merge = df_merge.merge(order_payments, on = 'order_id')
df_merge = df_merge.merge(sellers, on = 'seller_id')

df_merge.head()
```

# Merge du 1er et du second dataframe mergés

```python
df = df1.merge(df_merge, on = 'product_id')
df.info(verbose = True)
```

```python
df.head()
```

# Suppression de colonnes inutiles

colonnes sur les mensurations des produits

```python
df = df.drop(columns = ['product_weight_g',
                       'product_length_cm',
                       'product_height_cm',
                       'product_width_cm'])
```

```python
df.info(verbose = True)
```

# Suppression des doublons

```python
df.drop_duplicates(keep = 'first', inplace=True)
df.info(verbose = True) 
```

# Réagencement de la colonne 'order_purchase_timestamp'

```python
df['order_purchase_timestamp'].head(10)
```

Conversion de la colonne 'order_purchase_timestamps' au format datetime et création de colonnes 'year', 'month', 'day'
pour faciliter la navigation dans le dataset

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['year']= df['order_purchase_timestamp'].dt.year
df['month']= df['order_purchase_timestamp'].dt.month
df['day']= df['order_purchase_timestamp'].dt.day
# df['order_purchase_timestamp'].head(10)
df.info(verbose = True)
```

# Suppression des lignes avec valeur manquante

```python
df.dropna(axis = 0, how = 'any', inplace = True)
df.info(verbose = True)
```
